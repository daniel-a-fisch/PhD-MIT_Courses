# Stability in SUTVA

SUTVA (stable unit treatment value assumption) is the underlying assumption for Rubin's potential outcome framework. A violation of it makes causel inference much more difficult.

Stability of the treatment probably holds in...
- drug testing: in drug RCTs every participant in the treatment group receives the same pill, assuming a high production quality
- money transfers: an RCT that studies the effect of a \$ 1000 transfer can ensure that all participants receive the money either directly as cash or via a deposit
- A/B testing of new website: in a tech RCT, the code executed for every user is deterministic and identical and the treatment does not vary

All are highly regulated environments where the "dose" is controlled for every participant.

On the other hand, the stability assumption might be violated in...
- educational interventions: an RCT testing a new Maths curriculum. Different teachers might follow the new concepts in different ways and some might even skip parts and instead stick to the old curriculum. Therefore, students are not receiving the same treatment and we measure probably the treatment effect of some mixture of old and new curriculum
- a new surgical tool: a surgeon with 20 years of experience might use the tool differently compared to one who performs their first solo procedure. Moreover, there might be a learning affect adjusting to the new tool
- new massage technique: when evaluating the effectiveness of a new massage technique on professional athletes, the person treating might get tired over time and not execute the massage as well the more athletes have been treated before 

The treatment involves a human or a process or is a broad umbrella for multiple sub-treatments. Of course there are ways to mitigate those threats to the SUTVA assumption (e.g. some practice period with the new surgical tool, only a few massage techniques in a row or controlling for the time of massage) but is difficult to rule them out completely.
