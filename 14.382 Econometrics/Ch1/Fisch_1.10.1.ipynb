# 1.10.1 Partialling-out and adaptivity

### The Frisch-Waugh-Lovell Theorem (partialling-out)
The FWL theorem provides a fundamental result for regressions: the coefficient of a specific regressor in a  regression including many controls can be obtained by a "partialling-out" procedure.

Consider the model:
$Y=\beta D+\gamma W+\varepsilon$

where $Y$ is the log wage, $D$ is the gender indicator (the "treatment" of interest), and $W$ is a high-dimensional vector of controls (nuisance parameters). To estimate $\beta$ using partialling-out, we follow three steps:

1. Residualize $D$: Regress $D$ on $W$ and retrieve the residuals $\tilde{D}$. This represents the portion of gender variation that is independent of controls.

2. Residualize $Y$: Regress $Y$ on $W$ and retrieve the residuals $\tilde{Y}$. This represents the portion of wages that cannot be explained by controls.

3. Final Regression: Regress $\tilde{Y}$ on $\tilde{D}$. The resulting coefficient is algebraically identical to $\beta$ from the full regression.

As illustrated in the accompanying analysis of CPS wage data (restricting the sample to never-married individuals), both the full regression and the partialling-out procedure yield an identical gender coefficient of approximately -0.0696. This indicates that, holding all observed controls constant, women in this sample earn roughly 6.96% less than men.

### The adaptivity property

A critical theoretical feature of the partialling-out approach is its adaptivity property. It refers to the fact that the estimator for $\beta$ is "adaptive" to the estimation of the nuisance functions (the regressions of $Y$ on $W$ and $D$ on $W$). Specifically, small errors in estimating the nuisance parameters $\gamma_i$ (in the first stages) only result in second-order errors in the estimation of $\beta$. 

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the data (never-married workers from the CPS dataset)
file = "https://raw.githubusercontent.com/CausalAIBook/MetricsMLNotebooks/main/data/wage2015_subsample_inference.csv"
df = pd.read_csv(file)

df.describe()
```

Following some means to describe the sample.

```python
table = pd.DataFrame()

cols = ["lwage", "sex", "shs", "hsg", "scl",
        "clg", "ad", "ne", "mw", "so", "we",
        "exp1", "exp2", "exp3", "exp4",
        "occ", "occ2", "ind", "ind2"
        ]

table['Variable'] = ["Log Wage", "Sex", "Less then High School",
                     "High School Graduate", "Some College",
                     "College Graduate", "Advanced Degree",
                     "Northeast", "Midwest", "South", "West",
                     "Experience 1", "Experience 2", "Experience 3", "Experience 4",
                     "Occupation 1", "Occupation 2", "Industry 1", "Industry 2"
                    ]

table['All'] = df[cols].mean().values

table['Male'] = df[df['sex'] == 0][cols].mean().values
table['Female'] = df[df['sex'] == 1][cols].mean().values

# Show results
table
```

```python
print(table.loc[0, 'Female'] - table.loc[0, 'Male'])
```

In the sample, the difference of logwages between male and female never-merried workers is about -3.8%, despite women being relatively more educated.

Running an OLS regression only on the sex dummy confirms the result and provides a heteroscedastic standard error.

```python
nocontrol_fit = smf.ols("lwage ~ sex", data=df).fit()
nocontrol_est = nocontrol_fit.params['sex']
nocontrol_se = nocontrol_fit.HC3_se['sex']

print(f"The estimated sex coefficient is {nocontrol_est:.4f} "
      f"and the corresponding robust standard error is {nocontrol_se:.4f}.")
```

## Full regression
Running a regression with multiple controls, the predictive effect of sec can be obtained by looking only at its coeeficient. 

```python
control = "lwage ~ sex + (exp1+exp2+exp3+exp4)*(hsg+scl+clg+ad+C(occ2)+C(ind2)+mw+so+we)"

control_fit = smf.ols(control, data=df).fit()
control_est = control_fit.params['sex']
control_se = control_fit.HC3_se['sex']

print(f"The estimated sex coefficient is {control_est:.4f} "
      f"and the corresponding robust standard error is {control_se:.4f} (standard SE: {control_fit.bse['sex']:.4f}).")
```

Thus, when controlling for education, experience, occupation, industry, and region, women in this sample earn even 6.96% less than their male counterparts.

## Partialling out
Instead of running the full regression and selecting the coefficient of interest, one can use the Frisch-Waugh-Lovell theorem as follows to directly obtain the result:

* regress logwage on controls and residualize it (i.e. keep only the part of it which cannot be explained by the controls)
* regress sex on controls and residualize it
* regress the residulaized logwage on the residualized sex

The result is the predictive effect of sex on logwages net of the effect that can be explained by the controls.


```python
# Partialling-Out using ols

# models
flex_y = "lwage ~ (exp1+exp2+exp3+exp4)*(hsg+scl+clg+ad+C(occ2)+C(ind2)+mw+so+we)"  # model for log wages Y = gamma*W + u
flex_d = "sex ~ (exp1+exp2+exp3+exp4)*(hsg+scl+clg+ad+C(occ2)+C(ind2)+mw+so+we)"  # model for sex D = pi*W + v

# partialling-out the linear effect of W from Y (residualized log wage)
t_Y = smf.ols(flex_y, data=df).fit().resid
# partialling-out the linear effect of W from D (residualized sex)
t_D = smf.ols(flex_d, data=df).fit().resid

# regression of Y on D after partialling-out the effect of W
partial_fit = sm.OLS(t_Y, sm.add_constant(t_D)).fit()
partial_est = partial_fit.params[0]

# standard error
partial_se = partial_fit.HC3_se[0]

# confidence interval
print("95% CI: " + str(partial_fit.conf_int().values[0]))

print(f"The estimated sex coefficient is {partial_est:.4f} "
      f"and the corresponding robust standard error is {partial_se:.4f} (standard SE: {partial_fit.bse[0]:.4f}).")
```

As expected, the estimates coincide. However, even the standard SE are not exactly the same. Probaly manual partialling-out slightly underestimates the standard error because the second-stage regression doesn't know how many degrees of freedom were used up by the controls in the first stage.

```python
plt.figure(figsize=(8, 5))
sns.regplot(x=t_D, y=t_Y, scatter_kws={'alpha':0.3, 's':10}, line_kws={'color':'red'})
plt.title("Residualized Regression (Net of Controls)")
plt.xlabel("Residualized Sex")
plt.ylabel("Residualized Log Wage")
plt.show()
```

The plot shows the second stage regression after partialling out the ceffect of the controls. The slope of the resulting regression line visually represents the isolated wage penalty for college-educated women.
